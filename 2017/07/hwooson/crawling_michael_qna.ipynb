```python
import requests
from bs4 import BeautifulSoup
import pandas as pd


base_url = 'http://englishinkorean.com/questions/'
```

```python
resp = requests.get(base_url)
resp
```

```python
soup = BeautifulSoup(resp.text, 'html.parser')
```

```python
# 댓글 ol 엘리먼트
ol_cmt_list = soup.find('ol', 'comment-list')

# 대댓글이 있는 댓글만
cmts_depth1 = ol_cmt_list.select('li.comment.depth-1.parent')
print("대댓글이 있는 댓글 수: {}개".format(len(cmts_depth1)))
```

```python
# 마이클 선생님 대댓글 찾기
def find_micheal(cmt):
    state = False
    for auth in cmt.find_all('div', 'comment-author'):
        if 'Michael' in auth.text:
            state = True
    return state
```

```python
michael_cmt_sets = []

for cmt in cmts_depth1:
    if find_micheal(cmt):
        michael_cmt_sets.append(cmt)

print("마이클 선생님의 대댓글이 있는 댓글 수: {}개".format(len(michael_cmt_sets)))
```

```python
# 댓글 내용 파싱
def depth1_comment_parser(cmt):
    article = cmt.find('article').find('div', 'comment-content').text.strip()
    return article


# 마이클 댓글 파싱
def michael_comment_parser(cmt):
    children_cmts = cmt.find('ol', 'children').find_all('li')
    for item in children_cmts:
        if 'Michael' in item.find('div', 'comment-author').text:
            return item.find('article').find('div', 'comment-content').text.strip()
```

```python
qa_sets = []

for comment in michael_cmt_sets:
    question = depth1_comment_parser(comment)
    answer = michael_comment_parser(comment)
    item = {'question': question,
            'answer': answer}
    qa_sets.append(item)
```

```python
df = pd.DataFrame(qa_sets)
df
```

```python
writer = pd.ExcelWriter('michael_qna.xlsx')
df.to_excel(writer, 'sheet1')
writer.save()
```
